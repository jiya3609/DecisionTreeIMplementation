```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
print(iris)
```

```python
iris.target
```

We gotta know that there are total 3 classes in this classification problem

```python
import seaborn as sns
df=sns.load_dataset('iris')
df.head()
```

```python
X=df.iloc[:,:-1]
Y=iris.target
```

```python
#lets split the data now
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
treemodel=DecisionTreeClassifier()
```

```python
parameter={
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best','ransom'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']

}
```

```python
from sklearn.model_selection import GridSearchCV
cv=GridSearchCV(treemodel,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(X_train,Y_train)
```

```python
cv.best_params_
```

# **Let's generate the Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
treemodel=DecisionTreeClassifier(criterion='gini',max_depth=4,max_features='log2',splitter='best')
treemodel.fit(X_train,Y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel,filled=True)
```
